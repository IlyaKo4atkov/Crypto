```python
data = pd.read_csv('DOGE-USD.csv')
#Отображение данных датасета
data.head()
```

```python
#Проверка корреляции данных
data.corr(numeric_only=True)
```

```python
#Преобразование строковой даты и времени в правильный формат
data['Date'] = pd.to_datetime(data['Date'], infer_datetime_format=True)
data.set_index('Date', inplace=True)
data.isnull().any()
```

```python
#Проверка нулевых значений в датасете
data.isnull().sum()
```

```python
#Удаление пропущеных значений
data = data.dropna()
```

```python
#Проверка статистического анализа
data.describe()
```

```python
#Проведение анализа цены закрытия для дальнейшего прогнозирования
plt.figure(figsize=(20, 10))
x = data.groupby('Date')['Close'].mean()
x.plot(linewidth=2, color='r')
plt.xlabel('Дата', fontsize=20)
plt.ylabel('Значение', fontsize=20)
plt.title('Цена закрытия 2021 года', fontsize=20)
```

```python
#Проверка корреляции по закрытию цены
data["gap"] = (data["High"] - data["Low"]) * data["Volume"]
data["y"] = data["High"] / data["Volume"]
data["z"] = data["Low"] / data["Volume"]
data["a"] = data["High"] / data["Low"]
data["b"] = (data["High"] / data["Low"]) * data["Volume"]
abs(data.corr()["Close"].sort_values(ascending=False))
```

```python
#Оставляем только наиболее значимые значения
data = data[["Close", "Volume", "gap", "a", "b"]]
data.head()
```

```python
#Преобразование необходимых данных в массив данных и выделение тестовой и обучающей выборки
df2 = data.tail(30)
train = df2[:11]
test = df2[-19:]

print(train.shape, test.shape)
```

```python
#Разработка модели прогнозирования с помощью функции SARIMAX
from statsmodels.tsa.statespace.sarimax import SARIMAX
model = SARIMAX(endog=train["Close"], exog=train.drop(
    "Close", axis=1), order=(2, 1, 1))
results = model.fit()
print(results.summary())
```

```python
#Предсказание в виде временных рядов
start = 11
end = 29
predictions = results.predict(
    start=start,
    end=end,
    exog=test.drop("Close", axis=1))
predictions
```

```python
#Построение графика для визуализации
test["Close"].plot(legend=True, figsize=(12, 6))
predictions.plot(label='TimeSeries', legend=True)
```
